```python
#import the dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#loading the data
data = pd.read_csv('Obesity.csv')
data.info()
data = data.drop(columns='ID',axis=1)
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
contents = ['Gender', 'Label',]
lb = LabelEncoder()
for content in contents:
    data[content] = lb.fit_transform(data[content])

data
```

```python

#separating the target from features
x = data.drop(columns='Label', axis=1)
y = data['Label']
```

```python
# #data standardization
# from sklearn.preprocessing import StandardScaler
# scaler=StandardScaler()
# standard_data=scaler.fit_transform(x)
# X=standard_data
# X
```

```python
#splitting the data into train and test data
x_train,  x_test, y_train, y_test= train_test_split(x, y, test_size=0.15, stratify=y, random_state=3)
```

```python
model =  LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_train_pred = model.predict(x_train)
training_accuracy =  accuracy_score(x_train_pred, y_train)
training_accuracy

x_test_pred = model.predict(x_test)
test_accuracy =  accuracy_score(x_test_pred, y_test)
print('The training accuracy: ', training_accuracy)
print('The test accuracy: ', test_accuracy)
```

```python
#saving the trained model

import pickle
filename = 'trained_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
#loading the saved model
loaded_model = pickle.load(open('trained_model.sav', 'rb'))
```

```python
#making a predictive system
input_data = (25, 1, 175, 80,25.4)
input_array=np.asarray(input_data)
#reshape
input_array_reshape = input_array.reshape(1,-1)
prediction= loaded_model.predict(input_array_reshape)
print(prediction)

if (prediction == 0):
    print('The person is normal')
elif (prediction == 1):
    print('The person is Obese')
elif (prediction == 2):
    print('The person is Overweight')
else:
    print('The person is Underweight')




#1,25,Male,175,80,25.3,Normal Weight
```
